```python
def read_index_map(file_path=Cfg.INDEX_MAP_FILE):
    """Reads the sign to predict as json file."""
    with open(file_path, "r") as f:
        result = json.load(f)
    return result    

def read_train(file_path=Cfg.TRAN_FILE):
    """Reads the train csv as pandas data frame."""
    return pd.read_csv(file_path).set_index(Cfg.INDEX)

def read_landmark_data_by_path(file_path, input_root=Cfg.INPUT_ROOT):
    """Reads landmak data by the given file path."""
    data = pd.read_parquet(input_root / file_path)
    return data.set_index(Cfg.ROW_ID)

def read_landmark_data_by_id(sequence_id, train_data):
    """Reads the landmark data by the given sequence id."""
    file_path = train_data.loc[sequence_id]['path']
    return read_landmark_data_by_path(file_path)
```

# Visualization

```python
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands
mp_face_mesh = mp.solutions.face_mesh
mp_pose = mp.solutions.pose
```

```python
def get_random_sequence_id(train_data):
    idx = np.random.randint(0, len(train_data))
    return train_data.index[idx]
```

```python
def create_blank_image(height, width):
    return np.zeros((height, width, 3), np.uint8)

def draw_landmarks(
    data, 
    image, 
    frame_id, 
    landmark_type, 
    connection_type, 
    landmark_color=(255, 0, 0), 
    connection_color=(0, 20, 255), 
    thickness=1, 
    circle_radius=1
):
    """Draws landmarks"""
    df = data.groupby(['frame', 'type']).get_group((frame_id, landmark_type))
    landmarks = [landmark_pb2.NormalizedLandmark(x=lm.x, y=lm.y, z=lm.z) for idx, lm in df.iterrows()]
    landmark_list = landmark_pb2.NormalizedLandmarkList(landmark = landmarks)

    mp_drawing.draw_landmarks(
        image=image,
        landmark_list=landmark_list, 
        connections=connection_type,
        landmark_drawing_spec=mp_drawing.DrawingSpec(
            color=landmark_color, 
            thickness=thickness, 
            circle_radius=circle_radius),
        connection_drawing_spec=mp_drawing.DrawingSpec(
            color=connection_color, 
            thickness=thickness, 
            circle_radius=circle_radius))
    return image

def draw_left_hand(data, image, frame_id):
    return draw_landmarks(
        data, 
        image, 
        frame_id, 
        landmark_type='left_hand', 
        connection_type=mp_hands.HAND_CONNECTIONS,
        landmark_color=(255, 0, 0),
        connection_color=(0, 20, 255), 
        thickness=3, 
        circle_radius=3)

def draw_right_hand(data, image, frame_id):
    return draw_landmarks(
        data, 
        image, 
        frame_id, 
        landmark_type='right_hand', 
        connection_type=mp_hands.HAND_CONNECTIONS,
        landmark_color=(255, 0, 0),
        connection_color=(0, 20, 255),
        thickness=3, 
        circle_radius=3)

def draw_face(data, image, frame_id):
    return draw_landmarks(
        data, 
        image, 
        frame_id, 
        landmark_type='face', 
        connection_type=mp_face_mesh.FACEMESH_TESSELATION,
        landmark_color=(255, 255, 255),
        connection_color=(0, 255, 0))      
    
def draw_pose(data, image, frame_id):
    return draw_landmarks(
        data, 
        image, 
        frame_id, 
        landmark_type='pose', 
        connection_type=mp_pose.POSE_CONNECTIONS,
        landmark_color=(255, 255, 255),
        connection_color=(255, 0, 0),
        thickness=2, 
        circle_radius=2)

def create_frame(data, frame_id, frame_id_plot, height=1000, width=1000):
    image = create_blank_image(height, width)
    
    draw_pose(data, image, frame_id) 
    draw_left_hand(data, image, frame_id)    
    draw_right_hand(data, image, frame_id)  
    draw_face(data, image, frame_id)

    image = cv2.putText(img=image, text=f"{frame_id_plot:02}", org=(int(width*0.8), int(height*0.1)), 
                        fontFace=cv2.FONT_HERSHEY_SIMPLEX, fontScale=2, color=(255, 255, 255), 
                        thickness=5)
     
    return image
```

# Animation

```python
def create_frames(sequence_id, train_data, height=800, width=800):
    data = read_landmark_data_by_id(sequence_id, train_data)
    frame_ids = data['frame'].unique()
    images = [create_frame(data, frame_id=fid, frame_id_plot=i, height=height, width=width) for i, fid in enumerate(frame_ids)]
    return np.array(images)
```

```python
def create_animation(images, fig, ax):
    ax.axis('off')
    
    ims = []
    for img in images:
        im = ax.imshow(img, animated=True)
        ims.append([im])
    
    func_animation = animation.ArtistAnimation(
        fig, 
        ims, 
        interval=100, 
        blit=True,
        repeat_delay=1000)

    return func_animation

def get_sign_by_id(sequence_id, train_data):
    return train_data.loc[sequence_id]['sign']

def play_animation(sequence_id, train_data, height, width, figsize=(4, 4)):
    frames = create_frames(sequence_id, train_data, height=height, width=width)
    sign = get_sign_by_id(sequence_id, train_data)
    
    fig, ax = plt.subplots(1, 1, figsize=figsize)
    anim = create_animation(frames, fig, ax)
    ax.set_title(f'Sign: {sign}')
    
    video = anim.to_html5_video()
    html = display.HTML(video)
    display.display(html)
    plt.close()
```

# Check that preprocessing is OK

```python
# Landmarks
FACE_N_LANDMARKS = 468
POSE_N_LANDMARKS = 33
RHAND_N_LANDMARKS = 21
LHAND_N_LANDMARKS = 21
N_LANDMARKS = FACE_N_LANDMARKS + POSE_N_LANDMARKS + RHAND_N_LANDMARKS + LHAND_N_LANDMARKS # 543

# Contigous array access
FACE_OFFSET = 0
LHAND_OFFSET = FACE_N_LANDMARKS
POSE_OFFSET = LHAND_OFFSET + LHAND_N_LANDMARKS
RHAND_OFFSET = POSE_OFFSET + POSE_N_LANDMARKS

# Face landmarks groupings
to_include = {
    'FACE_REYE': [33, 7, 163, 144, 145, 153, 154, 155, 133, 173, 157, 158, 159, 160, 161, 246],
    'FACE_LEYE': [362, 382, 380, 381, 374, 373, 390, 249, 263, 466, 388, 387, 386, 385, 384, 398],
    'FACE_LIPS': [0, 37, 39, 40, 185, 61, 146, 91, 181, 84, 17, 314, 405, 321, 375,
             291, 409, 270, 269, 267, 13, 82, 81, 80, 191, 78, 95, 88, 178, 87,
             14, 317, 402, 318, 324, 308, 415, 310, 311, 312],
    'POSE_UPPER': list(range(0, 23)),
    'POSE_LOWER': list(range(23, 33))
}

def get_landmarks_to_include_idxs(landmarks_to_include):
    face = landmarks_to_include['face']
    pose = landmarks_to_include['pose']

    face_idxs = []
    if face == 'ALL':
        face_idxs = range(
            FACE_OFFSET, 
            FACE_OFFSET + FACE_N_LANDMARKS
        )
    else:
        for var in face:
            face_idxs.extend(to_include[var])
        face_idxs = [idx + FACE_OFFSET for idx in face_idxs]
    pose_idxs = []
    if pose == 'ALL':
        pose_idxs = range(
            POSE_OFFSET, 
            POSE_OFFSET + POSE_N_LANDMARKS
    )
    else:
        for var in pose:
            pose_idxs.extend(to_include[var])
        pose_idxs = [idx + POSE_OFFSET for idx in pose_idxs]

    face_idxs = np.array(sorted(face_idxs))
    pose_idxs = np.array(sorted(pose_idxs))
    rhand_idxs = np.array(
        range(
            RHAND_OFFSET, 
            RHAND_OFFSET + RHAND_N_LANDMARKS
        )
    )
    lhand_idxs = np.array(
        range(
            LHAND_OFFSET,
            LHAND_OFFSET + LHAND_N_LANDMARKS
        )
    )

    return np.concatenate((face_idxs, lhand_idxs, pose_idxs, rhand_idxs))
```

```python
prepdata = np.load('/home/gpu/Kaggle/GOOGLE-ASL/data/preprocessed/test/joint_data.npy')
prepidslbls = np.load('/home/gpu/Kaggle/GOOGLE-ASL/data/preprocessed/test/ids_lbls.npy')
prep_ids, prep_participant_ids, prep_lbls = prepidslbls[:, 0], prepidslbls[:, 1], prepidslbls[:, 2]
rawdata = read_train('/home/gpu/Kaggle/GOOGLE-ASL/data/raw/train.csv')

# Filter by the ids in the preprocessed data
rawdata = rawdata.loc[prep_ids]

rawseq = read_landmark_data_by_path(rawdata.iloc[0].path)
lmks_prep_data = get_landmarks_to_include_idxs({'pose': ['POSE_UPPER'], 'face': ['FACE_LIPS', 'FACE_REYE', 'FACE_LEYE']})
prepdata_idx = np.where(prep_ids == rawdata.iloc[0].name)[0][0]
prepseq = prepdata[prepdata_idx]
```

```python
def add_offset(row, frame_offset):
    frame = row.frame - frame_offset
    mult = N_LANDMARKS * frame
    lmk_idx = row.landmark_index
    if row.type == 'face':
        offset = FACE_OFFSET
    elif row.type == 'right_hand':
        offset = RHAND_OFFSET
    elif row.type == 'left_hand':
        offset = LHAND_OFFSET
    elif row.type == 'pose':
        offset = POSE_OFFSET
    else:
        raise ValueError
    lmk_idx += (offset + mult)
    
    return lmk_idx
    
def viz_original_and_prep_sequence(id, rawdata, lmks_prep_data):
    # Get original sequence
    rawseq_ref = rawdata.iloc[id]
    rawseq = read_landmark_data_by_path(rawseq_ref.path)
    rawseq_id = rawseq_ref.name
    # Get preprocessed sequence
    prepdata_idx = np.where(prep_ids == rawseq_id)[0][0]
    prepseq = prepdata[prepdata_idx]
    frame_offset = min(rawseq.frame.unique())
    n_frames = len(rawseq.frame.unique())
    all_ids = np.concatenate([lmks_prep_data + (N_LANDMARKS * i) for i in range(n_frames)])
    rawseq['prepid'] = rawseq.apply(add_offset, args=(frame_offset,), axis=1)
    # Assing preprocessed values to sequence
    rawseq.loc[rawseq.prepid.isin(all_ids), ['x','y','z']] = prepseq[:n_frames].reshape(-1, 3)
    # Replace 0.0 with NaN just for visualziation purposes
    rawseq.replace({'x': 0, 'y': 0, 'z': 0}, np.nan, inplace=True)
```

```python
# Increase the landmark_index by each type offset
def add_offset(row, frame_offset):
    frame = row.frame - frame_offset
    mult = N_LANDMARKS * frame
    lmk_idx = row.landmark_index
    if row.type == 'face':
        offset = FACE_OFFSET
    elif row.type == 'right_hand':
        offset = RHAND_OFFSET
    elif row.type == 'left_hand':
        offset = LHAND_OFFSET
    elif row.type == 'pose':
        offset = POSE_OFFSET
    else:
        raise ValueError
    lmk_idx += (offset + mult)
    
    return lmk_idx
    
frame_offset = min(rawseq.frame.unique())
n_frames = len(rawseq.frame.unique())
all_ids = np.concatenate([lmks_prep_data + (N_LANDMARKS * i) for i in range(n_frames)])
rawseq['prepid'] = rawseq.apply(add_offset, args=(frame_offset,), axis=1)
rawseq.tail()
```
